```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import requests
```

```python
url_list = []
for i in range(1,42):
    url = f"https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_6_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_6_0_na_na_na&as-pos=6&as-type=TRENDING&suggestionId=mobiles&requestId=7443f745-9634-4bb0-af43-57c2f2e29037&page={i}"
    url_list.append(url)
    
url_list[0]
```

```python
def create_html_page(url):
    uClient = urlopen(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = BeautifulSoup(page_html, features='html')
    return page_soup
```

```python
def create_final_data(page_soup):
    data = {"Product_Name":[],
            "Actual_price":[],
            "Sale_price":[],
            "Review":[],
            "Rattings":[],
            "Offer":[],
            "Camera":[],
            "Processor":[],
            "Battery":[],
            "Memory":[],
            "Display":[],
            "Avg_rating":[]}
    containers = page_soup.find_all("div",{"class":"tUxRFH"}) 
    for contain in containers:
        product_name = contain.find_all("div",{"class":"KzDlHZ"})[0].text
        data["Product_Name"].append(product_name)
        
        try:
            actual_price = contain.find_all("div",{"class":"yRaY8j ZYYwLA"})[0].text.replace(",","").replace("₹", "")
        except IndexError:
            actual_price = None        
        data["Actual_price"].append(actual_price)

        sale_price = contain.find_all("div",{"class":"Nx9bqj _4b5DiR"})[0].text.replace(",","").replace("₹", "")
        data["Sale_price"].append(sale_price)

        review_text = contain.find_all("span",{"class":"Wphh3N"})[0].text.replace(",","")
        parts = review_text.split("&")
        reviews = parts[1].strip().split(" ")[0]
        data["Review"].append(reviews)

        rating_text = contain.find_all("span",{"class":"Wphh3N"})[0].text.replace(",","")
        parts = rating_text.split("&")
        ratings = parts[0].strip().split(" ")[0]
        data["Rattings"].append(ratings)
        
        try:
            offer = contain.find_all("div",{"class":"UkUFwK"})[0].text
        except IndexError:
            offer = None        
        data["Offer"].append(offer)

        camera = contain.find_all("li",{"class":"J+igdf"})[2].text
        data["Camera"].append(camera)

        try:
            processor = contain.find_all("li", {"class": "J+igdf"})[4].text
        except IndexError:
            processor = None
        data["Processor"].append(processor)

        battery = contain.find_all("li",{"class":"J+igdf"})[3].text
        data["Battery"].append(battery)

        memory = contain.find_all("li",{"class":"J+igdf"})[0].text
        data["Memory"].append(memory)

        display = contain.find_all("li",{"class":"J+igdf"})[1].text
        data["Display"].append(display)

        avg_rating = contain.find_all("div",{"class":"XQDdHH"})[0].text
        data["Avg_rating"].append(avg_rating)
        
    return data
```

```python
data_frame_list = []
for i, url in enumerate(url_list):
    print(f"Started page {i}")
    final = create_html_page(url)
    df = create_final_data(final)
    data_frame_list.append(df)
```
